```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
df_org = pd.read_csv('/content/df_33_cleaned_filled_outlier_dummies.csv')
```

```python
df = df_org.copy()
```

```python
df.head(10)
```

```python
X = df.drop(['fuel_Diesel'], axis = 1)
y = df['fuel_Diesel']
```

**Train_Test_Split**

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 42)
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

**Modelling**

```python
from sklearn.linear_model import LogisticRegression
```

```python
logic = LogisticRegression()
```

```python
logic.fit(X_train_scaled, y_train)
```

```python
y_train_pred = logic.predict(X_train_scaled)
y_train_pred_prob = logic.predict_proba(X_train_scaled)
```

```python
y_pred = logic.predict(X_test_scaled)
y_pred_prob = logic.predict_proba(X_test_scaled)
```

```python
y_pred.shape, y_pred_prob.shape
```

```python
test_data = pd.concat([X_test, y_test], axis = 1)
test_data['pred'] = y_pred
test_data['pred_prob_1'] = y_pred_prob[:, 1]
test_data.sample(10)
```

**Model Performance on Classification Tasks**

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def eval_metric(model, X_train, y_train, X_test, y_test):
  y_train_pred = model.predict(X_train)
  y_pred = model.predict(X_test)

  print('Test_Set:')
  print(confusion_matrix(y_test, y_pred))
  print(classification_report (y_test, y_pred))
  print()
  print('Train_Set:')
  print(confusion_matrix(y_train, y_train_pred))
  print(classification_report(y_train, y_train_pred))
```

```python
eval_metric(logic, X_train_scaled, y_train, X_test_scaled, y_test)
```

**Cross Validation**

```python
from sklearn.model_selection import cross_validate
```

```python
score_logic_crval = cross_validate(logic, X_train_scaled, y_train,
                             scoring = ['precision', 'recall', 'f1', 'accuracy'], cv = 10)
score_logic_crval_df = pd.DataFrame(score_logic_crval, index = range(1, 11))
```

**Cross Validate for 0 class**

```python
from sklearn.metrics import make_scorer, precision_score, recall_score, f1_score
```

```python
precision_0 = make_scorer(precision_score, pos_label = 0)
recall_0 = make_scorer(recall_score, pos_label = 0)
f1_0 = make_scorer(f1_score, pos_label = 0)
```

```python
score_logic_0_crval = cross_validate(logic, X_train_scaled, y_train,
                               scoring = {'precision_for_0_class': precision_0, 'recall_for_0_class':recall_0,
                                          'f1_for_0_class': f1_0}, cv = 10)

score_logic_0_crval_df = pd.DataFrame(score_logic_0_crval, index = range(1, 11))
```

```python
score_logic_0_crval_df
```

```python
score_logic_0_crval_df.iloc[:, 2:].mean()
```

**GridSearchCV**

```python
from sklearn.model_selection import GridSearchCV
```

```python
penalty = ['l1', 'l2']
C = np.logspace(-1, 5, 5)
class_weight= ["balanced", None]
solver = ["lbfgs", "liblinear", "sag", "saga"]

param_grid = {'penalty': penalty, 'C': C,
              'class_weight': class_weight,
              'solver': solver}
```

```python
grid = GridSearchCV(logic, param_grid = param_grid, scoring = 'recall', n_jobs = -1)
```

```python
grid.fit(X_train_scaled, y_train)
```

```python
grid.best_params_
```

```python
eval_metric(grid, X_train_scaled, y_train, X_test_scaled, y_test)
```
