```python
def combine_cols(col1, col2):
  if pd.isnull(col1):
    if pd.notnull(col2):
      return col2
    else:
      return np.nan
  if pd.isnull(col2):
    if pd.notnull(col1):
      return col1
    else:
      return np.nan
  else:
    if col1 == col2:
      return col1
    else:
      return 'Conflict'
```

```python
def consumption_comb(a):
  if isinstance(a, list):
    a = [item for item in a if item !='\n']
    if isinstance(a[0], list):
      return a[0][0] if a[0] else np.nan
    else:
      return a[0]
```

```python
def consumption_city(a):
  if isinstance(a, list):
    a = [item for item in a if item !='\n']
    if len(a) == 3:
      if isinstance(a[1], list):
        return a[1][0] if a[1] else np.nan
      else:
        return a[1]
```

```python
def consumption_country(a):
  if isinstance(a, list):
    a = [item for item in a if item !='\n']
    if len(a) == 3:
      if isinstance(a[2], list):
        return a[2][0] if a[2] else np.nan
      else:
        return a[2]
```
